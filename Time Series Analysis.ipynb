```python
import pandas as pd
```

```python
# Load the dataset into a pandas DataFrame
df = pd.read_csv('city_day.csv')
```

```python
# Check the first few rows of the DataFrame
print(df.head())
```

```python
# Check the shape of the DataFrame
print("Shape of the DataFrame:", df.shape)
```

```python
# Inspect the column names
print("Column names:", df.columns)
```

```python
# Check the data types of each column
print("Data types of each column:")
print(df.dtypes)
```

```python
# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Set the 'Date' column as the index of the DataFrame
df.set_index('Date', inplace=True)
```

```python
pip install matplotlib
```

```python
import matplotlib.pyplot as plt

# Plot the time series data
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['AQI'], color='blue', linestyle='-')
plt.title('Air Quality Index (AQI) Over Time')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.grid(True)
plt.show()
```

This is a line graph showing AQI trends over time. 
Blue line: AQI values by date.
Insight: Observes changes in air quality over the period.

```python
# Calculate basic statistics of the AQI values
aqi_stats = df['AQI'].describe()
print("Basic Statistics of AQI:")
print(aqi_stats)
```

```python
# Resample the time series data to monthly frequency and calculate mean AQI for each month
monthly_mean_aqi = df['AQI'].resample('M').mean()

# Display the resampled data
print("Resampled Monthly Mean AQI:")
print(monthly_mean_aqi)
```

```python
# Plot the resampled monthly mean AQI data
plt.figure(figsize=(12, 6))
plt.plot(monthly_mean_aqi.index, monthly_mean_aqi.values, color='green', linestyle='-')
plt.title('Resampled Monthly Mean AQI Over Time')
plt.xlabel('Date')
plt.ylabel('Mean AQI')
plt.grid(True)
plt.show()
```

This graph displays the resampled monthly mean AQI over time.
Green line: Mean AQI values by month.
Insight: Tracks the average AQI trend month by month, offering insights into seasonal variations in air quality.

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Fill missing values in the 'AQI' column with the mean of the 'AQI' column
df['AQI_filled'] = df['AQI'].fillna(df['AQI'].mean())

# Perform seasonal decomposition
decomposition = seasonal_decompose(df['AQI_filled'], model='additive', period=365)

# Plot the decomposition results (same as previous code)
plt.figure(figsize=(12, 10))
plt.subplot(411)
plt.plot(df.index, df['AQI_filled'], label='Original', color='blue')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(df.index, decomposition.trend, label='Trend', color='red')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(df.index, decomposition.seasonal, label='Seasonal', color='green')
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(df.index, decomposition.resid, label='Residual', color='purple')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

This is a seasonal decomposition plot after filling missing values with the mean.
Blue line: Original AQI time series with missing values filled.
Red line: Trend component indicating long-term patterns.
Green line: Seasonal component representing seasonal fluctuations.
Purple line: Residual component showing random variation.
Insight: Despite filling missing values, underlying patterns, seasonality, and random fluctuations are still observable in the decomposed components.

```python
# Selecting a few cities for comparison
selected_cities = ['Delhi', 'Mumbai', 'Chennai', 'Kolkata']

# Create line plots for AQI values of selected cities
plt.figure(figsize=(12, 8))
for city in selected_cities:
    city_data = df[df['City'] == city]
    plt.plot(city_data.index, city_data['AQI_filled'], label=city)

plt.title('AQI Comparison Across Cities')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.legend()
plt.grid(True)
plt.show()
```

This code compares AQI values across selected cities with line plots.
Insight: Visualizes AQI trends in Delhi, Mumbai, Chennai, and Kolkata over time, aiding in city-wise air quality comparison.

```python
# Scatter plot of AQI vs. PM2.5
plt.figure(figsize=(10, 6))
plt.scatter(df['PM2.5'], df['AQI_filled'], color='orange', alpha=0.5)
plt.title('AQI vs. PM2.5')
plt.xlabel('PM2.5')
plt.ylabel('AQI')
plt.grid(True)
plt.show()
```

This code generates a scatter plot of AQI versus PM2.5 concentrations.
Orange dots represent individual data points.
Insight: Examines the relationship between PM2.5 and AQI, providing insights into air quality dynamics influenced by particulate matter.

```python
# Calculate rolling mean and rolling standard deviation
rolling_mean = df['AQI_filled'].rolling(window=30, min_periods=1).mean()
rolling_std = df['AQI_filled'].rolling(window=30, min_periods=1).std()

# Plot original AQI values and rolling statistics
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['AQI_filled'], label='Original', color='blue')
plt.plot(df.index, rolling_mean, label='Rolling Mean', color='red')
plt.plot(df.index, rolling_std, label='Rolling Std', color='green')

plt.title('Rolling Statistics: Mean and Standard Deviation')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.legend()
plt.grid(True)
plt.show()
```

This code calculates and plots the rolling mean and rolling standard deviation of AQI values.
Blue line: Original AQI values.
Red line: Rolling mean.
Green line: Rolling standard deviation.
Insight: Examines AQI trends with smoothed rolling mean and variability with rolling standard deviation over time.

```python
# Select only numeric columns for correlation analysis
numeric_columns = df.select_dtypes(include='number')

# Calculate correlations between AQI and other pollutants
correlation_matrix = numeric_columns.corr()

# Extract correlations with AQI
aqi_correlations = correlation_matrix['AQI_filled'].drop('AQI_filled')

# Print correlations
print("Correlations with AQI:")
print(aqi_correlations)
```

```python
# Calculate average AQI for each month
monthly_aqi = df.groupby(df.index.month)['AQI_filled'].mean()

# Plot average AQI for each month
plt.figure(figsize=(10, 6))
plt.plot(monthly_aqi.index, monthly_aqi.values, marker='o', linestyle='-')
plt.title('Average AQI by Month')
plt.xlabel('Month')
plt.ylabel('Average AQI')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

This code calculates and plots the average AQI for each month.
Marker: Circular markers represent average AQI for each month.
Insight: Visualizes monthly variations in AQI, highlighting potential seasonal patterns in air quality.
